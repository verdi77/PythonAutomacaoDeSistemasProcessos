# Automação de Sistemas e Processos com Python

### Desafio:

Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: seugmail+diretoria@gmail.com<br>
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

```python
import pyautogui # (O pyautogui vai automatizar o nosso mouse, teclado e tela do computador e sempre que importo pyautogui tenho q importar o pyperclip, tem na apostila)
import pyperclip # (Permite copiar um texto q está dentro do código e permite escrever caracteres que o pyautogui não reconhece como por exemplp ?)
import time

pyautogui.PAUSE = 1 # aqui estou dizendo: pyautogui, sempre q vc executar um comando espere por 1 segundo. Lembrando q o pyautogui.PAUSE funciona p todas as linhas.

# Passo 1 - Entrar no nosso sistema (entrar no link: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing)
pyautogui.hotkey("ctrl", "t")
link = "https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing"
pyperclip.copy(link)
pyautogui.hotkey("ctrl", "v")
pyautogui.press("enter")
# pyautogui.press("win") mostra a tela do windows. Press = quero apertar uma única tecla/hotkey = quero apertar várias teclas, combinação de teclas, quero duas ao mesmo tempo.
# hotkey significa atalho, combinação de teclas e coloco dentro do parenteses quais teclas quero apertar.
# No Python sempre q escreve-se um texto tem q colocar entre aspas simples '' o aspas duplas "".

# Passo 2 - Navegar no sistema (entrar na pasta Exportar)
time.sleep(5) #agora quero falar p Python q depois q der o enter, n quero mais q espere só o 1 segundo, quero q espere 5 segundos pq o site demora alguns segundos p carregar. O time.sleep controla o tempo. time.slee = espera 5 segundos, mas só que só neste lugar.
pyautogui.click(x=426, y=311, clicks=2) #agora vamos lá p baixo para descovrir a posição -> Use esse código para descobrir qual a posição de um item que queira clicar

# Passo 3 - Baixar o arquivo de vendas
time.sleep(2)
pyautogui.click(x=349, y=458) #clica no arquivo
pyautogui.click(x=1166, y=198) #clica nos tres pontinhos
pyautogui.click(x=1003, y=592) #manda fazer o dowload
time.sleep(3) # esperar ele fazer o download
```

### Vamos agora ler o arquivo baixado para pegar os indicadores

- Faturamento
- Quantidade de Produtos

```python
# Passo 4 - Calcular faturamento e quantidade de produtos vendidos
import pandas as pd # usar sempre q for importar dados, analisar tabelas

tabela = pd.read_excel(r"C:\Users\Adriana\Downloads\Vendas - Dez.xlsx") # colocar o r antes p ler exatamente o caminho da pasta e n dar conflito de caracteres.
faturamento = tabela["Valor Final"].sum()
quantidade = tabela["Quantidade"].sum()
display(tabela)
```

### Vamos agora enviar um e-mail pelo gmail

```python
# Passo 5 - Enviar o email para a diretoria
pyautogui.hotkey("ctrl", "t") # abre uma nova aba
link = "https://mail.google.com/"
pyperclip.copy(link)
pyautogui.hotkey("ctrl", "v")
pyautogui.press("enter")

# clicar no botão escrever
time.sleep(5)
pyautogui.click(x=114, y=206)

# escrever pra quem eu to mandando o email
pyautogui.write("pythonimpressionador+diretoria@gmail.com")
pyautogui.press("tab") # escolher o email
pyautogui.press("tab") # passar pro campo de assunto

# escrever o assunto
assunto = "Relatório de Vendas"
pyperclip.copy(assunto)
pyautogui.hotkey("ctrl", "v")
pyautogui.press("tab") # passar pro corpo do email

# escrever o corpo do email
texto = f"""
Prezados, bom dia

O faturamento foi de R${faturamento:,.2f}
A quantidade de produtos foi de {quantidade:,}

Abs
Lira Python"""
pyautogui.write(texto)

# enviar o email
pyautogui.hotkey("ctrl", "enter")

```
